<h1 id="Python:-comment-lire-un-fichier-texte">Python pour biochimistes: comment lire un fichier texte</h1>

<h2 id="Introduction-et-concepts-de-base">Introduction et concepts de base</h2>
<p><span style="font-size: 14pt;">Une des t&acirc;ches les plus courantes faite en bio-informatique est la lecture et l'&eacute;criture de fichiers en format texte. Les deux types de fichiers les plus utilis&eacute;s sont les fichiers de s&eacute;quences et les fichiers contenant des donn&eacute;es tabul&eacute;es (grilles de valeurs). Dans les exemples qui suivent sur cette page, nous allons nous concentrer sur les fichiers de s&eacute;quences.</span></p>
<p><span style="font-size: 14pt;">Un point des plus importants dans la lecture d'un fichier, quel qu'il soit est d'en connaitre la structure. Dans le cas d'un fichier de s&eacute;quence, la s&eacute;quence elle-m&ecirc;me n'est qu'un &eacute;l&eacute;ment de donn&eacute;e contenu dans le fichier; les autres &eacute;l&eacute;ments sont les m&eacute;ta-donn&eacute;es, c'est-&agrave;-dire les donn&eacute;es sur la donn&eacute;e. Il existe une multitude de formats de fichiers de s&eacute;quence (un petit aper&ccedil;u se trouve ici: <a href="https://emboss.sourceforge.net/docs/themes/SequenceFormats.html" target="_blank" rel="noopener">https://emboss.sourceforge.net/docs/themes/SequenceFormats.html</a>) mais celui qui sera surement celui que vous utiliserez le plus est le format FASTA.</span></p>
<p><span style="font-size: 14pt;">Cr&eacute;&eacute; pour le logiciel d'alignement du m&ecirc;me nom, ce format est tr&egrave;s simple en appliquant des &eacute;l&eacute;ments pr&eacute;sent&eacute;s plus t&ocirc;t.</span></p>
<ul>
<li><span style="font-size: 14pt;">La premi&egrave;re ligne d'une s&eacute;quence commencera toujours par le caract&egrave;re "&gt;" suivi d'une succession arbitraire d'informations sur une seule ligne, termin&eacute;e par un retour de chariot. Comme le contenu de cette ligne n'a pas de structure fixe, il peut &ecirc;tre difficile de trouver ce que l'on cherche. On peut cependant se baser sur les <a href="https://en.wikipedia.org/wiki/FASTA_format#NCBI_identifiers" target="_blank" rel="noopener">sp&eacute;cifications utilis&eacute;es par le NCBI</a> pour la cr&eacute;ation de leurs fichiers FASTA.</span></li>
<li><span style="font-size: 14pt;">Toutes les lignes qui suivront seront la s&eacute;quence, une lettre par nucl&eacute;otide ou par acide amin&eacute; (il faudra alors utilis&eacute; le code IUPAC &agrave; une lettre). &nbsp;Il n'y a pas de r&egrave;gles strictes sur la longueur des lignes (souvent, on voit de 50 &agrave; 80 caract&egrave;res/ligne) mais chaque ligne doit se terminer avec un retour de chariot.</span></li>
</ul>
<p><span style="font-size: 14pt;">En sachant cela. il est alors tr&egrave;s simple d'extraire les informations recherch&eacute;es.</span></p>

## Code de démonstration

<h3 id="&Eacute;tape-1:-Acqu&eacute;rir-le-fichier-en-le-pointant-dans-le-syst&agrave;me-de-fichiers">&Eacute;tape 1: Acqu&eacute;rir le fichier en le pointant dans le syst&egrave;me de fichiers</h3>
<p>On peut le faire de plusieurs fa&ccedil;ons diff&eacute;rentes mais ici, on se limitera &agrave; deux: statique et interactive via le m&eacute;thode <em>input()</em>.</p>

```python
# Chargeons les librairies nécessaires
import os.path
import sys

#
# Méthode interactive
# Le script attendra l'entrée faite par l'usager
#
seq = input("PATH vers fichier? :")

#
# Il faut trouver ce que "~" veut dire... La méthode expanduser "interprete"
# que "~" veut dire "/Users/foisys" sur Mac OS X :-)
#
seq = os.path.expanduser(seq)
```

Maintenant, on va en lire le contenu:

```python
#
# La fonction open()permettra de lire le fichier pour le mettre dans un objet
# simplement appelé file
#
# La méthode readlines() permettra la lecture de toutes les lignes dans la
# variables allLines sous la forme d'une liste, une ligne par item de la liste
#
with open(seq,'r') as file:
    allLines = file.readlines()
#
# Remarque: allLines est une liste d'éléments, ici chacune des lignes du fichier
#
print("Nature de allLines: "+str(type(allLines)))
#
# On affiche avec la fonction print()
#
# Remarque: observez les crochets carrés au début ("[") et à la fin ("]")
#
print("Contenu de "+ allLines[0])
print(allLines)
```

<h3 id="&Eacute;tape-2:-Rendre-&ccedil;a-plus-lisible...">&Eacute;tape 2: Rendre &ccedil;a plus lisible...</h3>
<p>Bon, on a lu le fichier mais ce n'est pas tr&egrave;s lisible... La variable <em>allLines</em> contient tout le texte contenu dans le fichier, y compris les retours de chariots. On pourrait remplacer tous ces carat&egrave;res par autre chose mais &ccedil;a, c'est pour un autre cours :-) Utilisons le fait que allLines est une liste qui poss&egrave;de une succession d'items identiques (des lignes de caract&egrave;res) qui sont index&eacute;s dans l'ordre de la lecture des lignes pour n'&eacute;crire que les 10 premi&egrave;res lignes de la s&eacute;quence.</p>

```python
# Chargeons la librairie nécessaire
import os.path
import sys
#
# Il faut trouver ce que "~" veut dire... La méthode expanduser "interprete"
# que "~" veut dire "/Users/foisys" sur Mac OS X :-)
#
seq = os.path.expanduser(seq)
#
# Compteur utilisé plus bas
i = 0

# On crée la pointeur vers le fichier
with open(seq, 'r') as file:
    #
    # Tant et aussi longtemps que le fichier contientdra une ligne
    #
    for i in range(11):
        print(allLines[i].strip())
    print("blablabla...")
```

<h3 id="&Eacute;tape-3:-Extraire-la-s&eacute;quence-du-reste-du-fichier---M&eacute;thode-A.">&Eacute;tape 3: Extraire la s&eacute;quence du reste du fichier.</h3>

```python
#
# Initialisons une variable vide
#
fullseq = ""
#
# On se dirige vers le fichier
#
with open(seq, 'r') as file:
    #
    # Tant et aussi longtemps que allLines contiendra un item
    # Rappel: allLines est une liste d'items
    #
    for aLine in allLines:
        #
        # Si la ligne ne contient pas le caractère ">"
        # La valeur -1 spécifie que ">" n'existe pas sur la ligne
        #
        if aLine.find('>')==-1:
          #
          # On ajoute la ligne à la variable fullseq, qui se trouve allongée à chaque ligne
          #
          fullseq = fullseq+aLine.strip()
print("Impression de la sequence:")
#
# Prenez note: l'affichage se fait sur plusieurs lignes mais en fait, 
# ce n'est qu'une seule ligne que la fenêtre "replie" pour la faire 
# entrer dans l'espace
#
print(fullseq)
```

```python
#
# Comme fullseq est une simple chaine de caractères, on peut utiliser la
# méthode len() pour en connaitre le nombre de bases/acides aminés
#
print("Taille de la sequence:")
print(len(fullseq))
```
