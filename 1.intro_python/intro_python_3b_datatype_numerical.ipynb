<h2 id="Types-de-valeurs-num&eacute;riques">Types de valeurs num&eacute;riques</h2>
<p>Dans plusieurs languages de programmation, une variable num&eacute;rique doit &ecirc;tre construite en tenant compte de certaines caract&eacute;ristiques; par exemple, est-ce un nombre entier ou r&eacute;el. De la m&ecirc;me mani&egrave;re, il faut parfois pr&eacute;ciser les balises extr&egrave;mes; par exemple, en Java, une valeur r&eacute;elle peut &ecirc;tre de type <em>float</em> ou de type <em>double</em>, la diff&eacute;rence &eacute;tant que le premier (32 bits) est moins grand que le deuxi&egrave;me (64 bits) et occupe donc moins de place en m&eacute;moire. En Python, il n'existe que des valeurs, enti&egrave;res ou r&eacute;elles, &agrave; 64 bits. Ceci &eacute;tant dit, est-il possible de convertir l'un en l'autre ou <em>vice versa</em>? Sans probl&egrave;me: via la notion de conversion explicite (<em>type casting</em> en langue de Shakespeare). Exceptionnellement en Python, il existe une troisi&egrave;me type, <em>complex</em>, qui repr&eacute;sente les nombres complexes (duh!). Pour ce type de nombre, aucune conversion n'est possible.</p>
<h3>Exemples:</h3>

```python
i = 5
j = 3.141592653589793238
print(type(i))
print(i)
print(type(j))
print(j)
```

```python
#
# Exemples et effets de la conversion
#
print(float(i)) # Remarquez l'ajout du zéro pour représenter une valeur réelle
print(int(j)) # Remarquez la troncation à la seule valeur entière
```

```python
# Si vous voulez limiter la précision d'uen valeur réelle
# utiliser la fonction round()
# Ici, k demeure uen valeur réelle
j = 3.141592653589793238
k = round(j,4)
#
print(type(k))
print(k)
# Une autre méthode est l'utilisation de la méthode format() sur la sortie
# Dans ce cas, m deviendra une chaine de caractères
m = format(j,".6f")
#
print(type(m))
print(m)
```

<h2>Op&eacute;rations possibles sur des variables num&eacute;riques</h2>
<p>Il est &eacute;videmment possible de faire toute une vari&eacute;t&eacute; d'op&eacute;rations sur des variables num&eacute;riques :-) Les plus utilis&eacute;es sont les op&eacute;rations arithm&eacute;tiques et les op&eacute;rations de comparaison/logique.</p>
<p>Une op&eacute;ration arithm&eacute;tique est tr&egrave;s simple:</p>
<p>&nbsp;</p>

```python
#
# Définissons deux variables
#
i = 3
j = 3
#
# Addition
#
k = i + j
print(k)
#
# Soustraction
#
k = i - j
print(k)
#
# Multiplication
#
k = i * j
print(k)
#
# Division
#
k = i / j
print(k)
#
# Modulo (reste d'une division)
#
k = i % j
print(k)
#
# Puissance
#
k = i ** j
print(k)
```

<p>Une extension des ces op&eacute;rations sont les op&eacute;rations d'incr&eacute;mentation (tr&egrave;s utiles pour la cr&eacute;ation de compteurs ):</p>

```python
#
# Définissons deux variables
#
i = 3
j = 3
#
# On prend l'opérateur arithmétique suivi du symbole "="
#
# Ajouter 1 de la valeur j
i += 1
print(i)
# Soustraire 1 de la valeur j
j-=1
print(j)
# Multiplier i par 2
# Remarquez que i = 4 maintenant :-)
i*=2
print(i)
```

<p>Une op&eacute;ration logique ne peut retourner que deux valeurs possibles: VRAI ou FAUX. Ces op&eacute;rations sont critiques dans le fonctionnement des boucles d'ex&eacute;cutions (if..., for..., while...)</p>

```python
#
# Définissons deux variables
#
i = 3
j = 4
print(i == j)
print(i < j)
print(i > j)
print(i <= j)
print(i >= j)
print(i != j)
```
