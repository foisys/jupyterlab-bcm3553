<h1 id="Python-pour-biochimistes:-d&eacute;finitions-des-variables">Python pour biochimistes: d&eacute;finitions des variables booléennes</h1>



<h2>Qu'est-ce qu'une variable bool&eacute;&eacute;enne?</h2>
<p>Tr&egrave;s simplement, c'est une variable qui ne peut avoir que deux valeurs : TRUE (vrai) ou FALSE (faux). Et &agrave; quoi est-ce utile? Dans un tas de circonstance, une s&eacute;rie d'actions ne se fera que si et seulement si une ou plusieurs conditions sont respect&eacute;es; un autre exemple serait plut&ocirc;t que cette s&eacute;rie se fasse tant et aussi longtemps que les m&ecirc;mes conditions sont respect&eacute;es. On verra dans quelques pages comment on utilise tout &ccedil;a :-)</p>
<p>Un exemple:</p>

```python
# On veut avoir une suite de 0 à 10
for i in range(11):
  # Remarquez le ==
  # Selon la situation, il faut que 
  # le contenu de la condition retourne TRUE ou FALSE
  if(i%2==0):
    print(i)
```

<h2>Jouez avec les conditions!</h2>
<p>Comment manipuler tout &ccedil;a? On peut sp&eacute;ficier l'&eacute;avaluation de la condition selon qu'on veut avoir une VRAI ou un FAUX</p>
<ul>
<li>Si on veut tester notre condition avec un VRAI, on utilise la double &eacute;galit&eacute; (==);</li>
<li>So on veut tester avec un FAUX, on utilise != qui sera &eacute;valuer comme "n'est pas &eacute;gal &agrave;"</li>
<li>Dans l'&eacute;valuation de conditions num&eacute;riques, on peut utiliser: "&gt;" et "&lt;","&gt;=" et "&lt;="</li>
</ul>
<p>Est-il possible de mettre plusieurs conditions &agrave; resepecter dans une seule &eacute;valuation? Oui! Contrairement &agrave; des languages comme C++ ou Java, qui utilise des syntaxes particuli&egrave;res comme le ET logique ("&amp;&amp;") ou le OU logique ("||"), on utilise sp&eacute;cifiquement les mots cl&eacute; OR et AND en minuscules pour combiner des conditions. &Ccedil;a peut mener &agrave; des situations assez complexes et il peut-etre utile d'utiliser des parenth&egrave;ses pour s&eacute;parer les conditions &agrave; &eacute;valuer.&nbsp;</p>
<p>En exemple:</p>

```python
# On veut avoir une suite de 0 à 10
for i in range(11):
  # Remarquez le ==
  # Selon la situation, il faut que 
  # le contenu de la condition retourne TRUE ou FALSE
  if(i%2==0 and i>=5):
    print(i)
```
