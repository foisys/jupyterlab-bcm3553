<h1 id="Python-pour-biochimistes:-bien-&eacute;crire-votre-code">Python pour biochimistes: bien &eacute;crire votre code</h1>
<h2 id="Introduction">Introduction</h2>
<p>&Eacute;videmment, un script contiendra plusieurs items, r&eacute;parti sur plusieurs lignes. Comment Python interpr&egrave;te vos instructions? Dans une majorit&eacute; de languages, les blocs d'instructions sont s&eacute;par&eacute;s par des caract&egrave;res sp&eacute;ciaux, tel que les acoclades ('{' ou '}'). Python fonctionne diff&eacute;remment: on utilisera plut&ocirc;t des indentations, c'est-&agrave;-dire des espaces qui s&eacute;parerons les blocs.</p>
<p>Note: l'indentation par d&eacute;faut de Jupyter est de 4 espaces.</p>
<h2>Comment bien faire les indentations</h2>
<p>Un exemple:&nbsp;</p>

```python
#
# La fonction range() permet de créer une itération, entre 0 et 5
# dans notre cas
#
for i in range(6):
    # Remarquez que il faut forcer i et j à devenir une chaine de caractères
    # pour que la concaténation fonctionne
    print("Boucle "+ str(i))
    j = i+10
    print("Valeur: "+str(j))
print("Terminé!")    
```

<h2>Quand &ccedil;a ne fonctionne pas...</h2>
<p>Que se passe-t-il si on n'a pas la bonne indentation? Ajoutons une indentation &agrave; notre 2&egrave;me ligne&nbsp;<em>print()</em>.</p>
<p>Exemple:</p>

```python
for i in range(6):
    print("Boucle "+ str(i))
    j = i+10
        print("Valeur: "+str(j))
```

<p>Remarquez que la ligne "Termin&eacute;!" ne s'affiche que lorsque les instructions de la boucle <em>for</em>, qui ont la m&ecirc;me indentation, ont &eacute;t&eacute; ex&eacute;cut&eacute;es. &nbsp;Plus d'informations sur les boucles &agrave; venir 😃</p>
