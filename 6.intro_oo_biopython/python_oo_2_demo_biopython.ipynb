<h1 id="Python-pour-biochimistes:-comment-lire-un-fichier-texte,-la-voie-BioPython">Python pour biochimistes: comment lire un fichier texte, la voie BioPython</h1>
<p>Le projeet <a href="https://www.biopython.org" target="_blank" rel="noopener">BioPython</a> cherche &agrave; cr&eacute;er des outils d'analyse des s&eacute;quences biologiques pour simplifier l'utilisation de Python pour des programmeurs du dimanche que sont la majorit&eacute; des biochimistes! Le projet regroupe un ensemble impressionant de modules pour r&eacute;aliser les t&acirc;ches usuelles et m&ecirc;me plus qui sont construits en utilisant le mod&egrave;le orioent&eacute; objet.</p>
<p>Par exemple, lire les fichiers de s&eacute;quence dans divers formats (et m&ecirc;me des fichiers qui ne sont pas des fichiers de s&eacute;quence!!) devient tr&egrave;s ais&eacute;. Le module SeqIO permet la lecture automatique d'un fichier et retourne un objet de type SeqRecord. Ce dernier contient une repr&eacute;sentation standardis&eacute;e des infos contenus dans le fichier pour des analyses ult&eacute;rieures. On acc&eacute;de aux champs d'infos avec diff&eacute;rentes m&eacute;thodes, qui peuvent vari&eacute;es d'un format &agrave; l'autre.</p>

```python
# Comme pour l'exemple précédent
import os.path as path
# Importons de la librairie BioPython
# le module nécessaire
from Bio import SeqIO

#
# Méthode interactive
# Le script attendra l'entrée faite par l'usager
#
seq = input("PATH vers fichier? :")

#
# Il faut trouver ce que "~" veut dire... La méthode expanduser "interprete"
# que "~" veut dire "/Users/foisys" sur Mac OS X :-)
#
seq = path.expanduser(seq)

with open(seq,'r') as file:
    #
    # Remplacer fasta par gb et tester avec le fichier
    # ../z.misc_files/h.sapiens.cttn.iso.a.cdna.gb
    #
    for record in SeqIO.parse(file, "fasta"):
        print(record)
```

```python
print("Impression de la sequence:")
print(record.seq)
```

```python
print("Sequence ID:")
print(record.id)
```

```python
print("Taille de la sequence:")
print(len(record.seq))
```

```python
print("Features de la sequence:")
list = record.features
for i in list:
    print(i)
```
