<h1 id="Python-pour-biochimistes:-comment-lire-un-fichier-texte,-pareil-mais-diff&eacute;rent">Python pour biochimistes: comment lire un fichier texte, pareil mais diff&eacute;rent</h1>

<h2 id="Introduction-et-concepts-de-base">Introduction et concepts de base</h2>
<p><span style="font-size: large;">Ok, on a lu un fichier texte mais &eacute;tait-ce vraiment un fichier FASTA? Comment s'en assurer? En v&eacute;rifiant que la premi&egrave;re ligne du fichier contient le caract&egrave;re '&gt;'. L'exemple suivant montre &eacute;galement qu'il n'est pas n&eacute;cessaire de lire tout le fichier d'un trait.</span></p>

## Code de démonstration

<h3 id="&Eacute;tape-1:-Acqu&eacute;rir-le-fichier-en-le-pointant-dans-le-syst&agrave;me-de-fichiers">&Eacute;tape 1: Acqu&eacute;rir le fichier en le pointant dans le syst&egrave;me de fichiers</h3>
<p>On commence par lire un fichier:</p>

```python
# Chargeons les librairies nécessaires
import os.path
import sys
#
# Méthode interactive
# Le script attendra l'entrée faite par l'usager
#
seq = input("PATH vers fichier? :")
#
# Il faut trouver ce que "~" veut dire... La méthode expanduser "interprete"
# que "~" veut dire "/Users/foisys" sur Mac OS X :-)
#
seq = os.path.expanduser(seq)
```

<h3 id="&Eacute;tape-1:-Acqu&eacute;rir-le-fichier-en-le-pointant-dans-le-syst&egrave;me-de-fichiers">&Eacute;tape 2: N'affichez que le d&eacute;but du fichier mais tout de m&ecirc;me lire toute la s&eacute;quence</h3>
<p>En plus, on v&eacute;rifie le format du fichier selon notre standard FASTA. Comme nous testons la premi&egrave;re ligne, on n'a pas &agrave; lire le fichier en entier via la fonction <em>readlines()</em>. &Ccedil;a sauve du temps si on a un grand fichier ;-)</p>

```python
#
# Compteur utilisé ci-dessous
#
i = 0
#
# Variable vide à remplir
#
fullseq = ""
# On crée la pointeur vers le fichier
with open(seq, 'r') as file:
    #
    # Ici, on lit le fichier ligne par ligne plutôt que d'un trait
    #
    # Tant et aussi longtemps que le fichier contiendra une ligne
    #
    for aLine in file:
      if(i<=11):
        #
        # On sait que le format FASTA spécifie que la 1ère ligne doit commencer 
        # par un ">", donc si on ne trouve pas ce caractère dans la 1ère ligne lue...
        #
        if(i==0 and aLine.find(">")==-1):
          # 
          # On force l'arrêt du script
          #
          sys.exit("Ce fichier n'est pas en format FASTA!! Vérifiez-le :-)")
        else:
          print(aLine.strip())
          # Ajoutons 1 à i à chaque ligne lue
          # Pour ne pas tout affiché... On affiche seulement 10 lignes de sequence
          i+=1
      #
      # On veut remplir la variable fullseq avec la séquence uniquement, sans la 1ère
      # ligne
      #
      if aLine.find(">")==-1:
        fullseq = fullseq+aLine.strip()
    print("blablabla...")
```

<h3 id="&Eacute;tape-4:-Extraire-la-s&eacute;quence-du-reste-du-fichier.">&Eacute;tape 2: V&eacute;rifier que nous avons tout de m&ecirc;me lu toute la s&eacute;quence...</h3>

```python
print("Impression de la sequence:")
#
# Prenez note: l'affichage se fait sur plusieurs lignes mais en fait, 
# ce n'est qu'une seule ligne que la fenêtre "replie" pour la faire 
# entrer dans l'espace
#
print(fullseq)
```

```python
#
# Comme fullseq est une simple chaine de caractères, on peut utiliser la
# méthode len() pour en connaitre le nombre de bases/acides aminés
#
print("Taille de la sequence:")
print(len(fullseq))
```
