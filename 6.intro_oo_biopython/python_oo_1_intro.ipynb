<p>La programmation orient&eacute;e-objet est un sujet complexe mais qui dans la mise en action au quotidien est relativement simple. C'est "<span style="color: #202122; font-family: sans-serif; font-size: 14px; background-color: #ffffff;">consiste en la d&eacute;finition et l'interaction de briques logicielles appel&eacute;es&nbsp;</span><em style="color: #202122; font-family: sans-serif; font-size: 14px;"><a style="text-decoration: none; color: #795cb2; background: none; overflow-wrap: break-word;" title="Objet (informatique)" href="https://fr.wikipedia.org/wiki/Objet_(informatique)" target="_blank" rel="noopener">objets</a></em><span style="color: #202122; font-family: sans-serif; font-size: 14px; background-color: #ffffff;"> ; un objet repr&eacute;sente un concept, une id&eacute;e ou toute entit&eacute; du monde physique, comme une voiture, une personne ou encore une page d'un livre. Il poss&egrave;de une structure interne et un comportement, et il sait interagir avec ses pairs. Il s'agit donc de repr&eacute;senter ces objets et leurs relations&nbsp;; l'interaction entre les objets via leurs relations permet de concevoir et r&eacute;aliser les fonctionnalit&eacute;s attendues, de mieux r&eacute;soudre le ou les probl&egrave;mes. D&egrave;s lors, l'&eacute;tape de mod&eacute;lisation rev&ecirc;t une importance majeure et n&eacute;cessaire pour la POO. C'est elle qui permet de transcrire les &eacute;l&eacute;ments du r&eacute;el sous forme virtuelle.</span>" (Wikip&eacute;dia)</p>
<p>En pratique?&nbsp;</p>
<ul>
<li>Des objets aux propri&eacute;t&eacute;s d&eacute;finies sont cr&eacute;&eacute;s par une instantiation d'une classe.</li>
<li>La classe contient les m&eacute;thodes ou fonctions concernant ce type d'objets</li>
<li>Le prgrammeur interagit avec les propri&eacute;t&eacute;s de l'objet en invoquant les m&eacute;thodes de cette classe.</li>
</ul>
<p>Un exemple: d&eacute;finissons une classe caract&eacute;risant les cercles.</p>

```python
class Cercle:
    # Pour avoir une valeur de pi
    import math
    #
    # Constructeur nécessaire à l'instantiation d'un objet de la classe
    #
    # Le mot-clé self est nécessaire pour spécifier les propriétés de l'objet
    # créé
    #
    # Si on ne met que le mot-clé self, on déclare seulement Cercle() 
    #
    def __init__(self, radius):
        self.radius = radius
    
    def getArea(self):
        self.area = self.math.pi * self.radius**2
        return self.area
    
    def getPerim(self):
        self.perim = 2* self.math.pi * self.radius
        return self.perim

#
# Créons un cercle
# 
unCercle = Cercle(10)
#
# Obtenons les valeurs avec les méthodes de l'objet Cercle
#
print("Surface de unCercle: "+str(unCercle.getArea()))
# Comment limiter le nombre de chiffres suivant le point :-)
x= "{:.4f}".format(unCercle.getArea())
print("Surface de unCercle (arrondie):"+x)
print("Perimetre de unCercle: "+str(unCercle.getPerim()))
x= "{:.4f}".format(unCercle.getPerim())
print("Perimetre de unCercle (arrondi):"+x)
```
