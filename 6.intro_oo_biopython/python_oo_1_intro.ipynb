<p>En pratique?&nbsp;</p>
<ul>
<li>Des objets aux propri&eacute;t&eacute;s d&eacute;finies sont cr&eacute;&eacute;s par une instantiation d'une classe.</li>
<li>La classe contient les m&eacute;thodes ou fonctions concernant ce type d'objets</li>
<li>Le prgrammeur interagit avec les propri&eacute;t&eacute;s de l'objet en invoquant les m&eacute;thodes de cette classe.</li>
</ul>
<p>Un exemple: d&eacute;finissons une classe caract&eacute;risant les cercles.</p>

```python
class Cercle:    
#    import math
    pi = 3.14159
    # Constructeur
    def __init__(self, radius):
        self.radius = radius
    
    def getArea(self):
        self.area = self.pi * self.radius**2
        return self.area
    
    def getPerim(self):
        self.perim = 2* self.pi * self.radius
        return self.perim

#
# Créons un cercle
# 
unCercle = Cercle(10)
print(unCercle.getArea())
print(unCercle.getPerim())
```
