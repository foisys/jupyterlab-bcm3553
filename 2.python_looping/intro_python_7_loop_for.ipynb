<h1 id="Python-pour-biochimistes:-les-boucles">Python pour biochimistes: les boucles</h1>
<h2 id="Introduction">Introduction</h2>
<p>Les boucles sont des &eacute;l&eacute;ments incontornables dans le processus de traitement des donn&eacute;es dans le contexte de l'ex&eacute;cution d'un script. Comme tout autre language, Python poss&egrave;de divers m&eacute;canismes permettant d'ex&eacute;cuter (ou pas) des t&acirc;ches:</p>
<ul>
<li>Les boucles <em>if</em>;</li>
<li>Les boucles <em>for</em>;</li>
<li>Les boucles <em>while</em>.</li>
</ul>
<h2 id="Les-boucles-If">Les boucles For ... in ...</h2>
<p>Une autre utilisation des boucles est l'it&eacute;ration sur un ensemble de donn&eacute;es qui seront trait&eacute;es de mani&egrave;re identiques ou diff&eacute;renci&eacute;es, par exemple: l'&eacute;criture d'un rapport. C'est l&agrave; que la commande <em>for</em> sera tr&egrave;s utile; de plus, elle utilise souvent les it&eacute;rateurs implicites qui se trouvent derri&egrave;re certains types de variables.</p>

```python
#
# Prenons un intervalle entre 1 et 9
#
for i in range(9):
    print(i)
    if (i%2==0):
        print("le nombre i est pair")
    else:
        print("le nombre i est impair")
```

<p>&Agrave; tout moment, &agrave; l'atteinte d'une certaine condition, une boucle <em>for</em>, comme pour une boucle while, peut &ecirc;tre stopp&eacute;e gr&acirc;ce &agrave; la commande <em>break</em>:</p>

```python
#
# Prenons un intervalle entre 1 et 9
#
for i in range(9):
    if (i > 5):
        break 
    print(i)
    if (i%2==0):
        print("le nombre i est pair")
    else:
        print("le nombre i est impair")
```
