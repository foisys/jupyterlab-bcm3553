<h1 id="Python-pour-biochimistes:-les-boucles">Python pour biochimistes: les boucles</h1>
<h2 id="Introduction">Introduction</h2>
<p>Les boucles sont des &eacute;l&eacute;ments incontornables dans le processus de traitement des donn&eacute;es dans le contexte de l'ex&eacute;cution d'un script. Comme tout autre language, Python poss&egrave;de divers m&eacute;canismes permettant d'ex&eacute;cuter (ou pas) des t&acirc;ches:</p>
<ul>
<li>Les boucles <em>if</em>;</li>
<li>Les boucles <em>for</em>;</li>
<li>Les boucles <em>while</em>.</li>
</ul>
<h2 id="Les-boucles-If">Les boucles If</h2>
<p>Tr&egrave;s souvent, on devra rendre l'ex&eacute;cution d'une suite d'op&eacute;rations <strong>conditionnelle</strong> au respect d'une condition qui sera &eacute;valu&eacute;e sous la forme d'une op&eacute;ration bool&eacute;enne. Par exemple: si x &gt; i, alors on fait quelque chose; si non, on ne fera rien.</p>

```python
# Méthode interactive
# Le script attendra l'entrée faite par l'usager
#
i = input("Entrer un entier i entre 0 et 9 :")
# Pourquoi le typecasting?
if (int(i)<5):
    print("i est inférieur à 5")

# On peut mettre plus de conditions avec les mots clé "and" ou "or"
if (int(i)<5 and int(i)%2 == 0):
    print("i est inférieur à 5 et est pair")
```

<h2 id="Python-pour-biochimistes:-les-boucles">Les boucles If ... Else</h2>
<p>Ces boucles sont une extension du concept de la boucle conditionnelle <em>if</em>: si la condition est respect&eacute;e, faite A, si non (<em>else</em>), faite B.<em>&nbsp;</em></p>

```python
# Méthode interactive
# Le script attendra l'entrée faite par l'usager
#
i = input("Entrer un entier i entre 0 et 9 :")
# Pourquoi le typecasting?
if (int(i)<5):
    print("i est inférieur à 5")
else:
    print("i est supérieur ou égal à 5")
```

<h2>Les boucles If...elif...else</h2>
<p>On peut faire un suite d'&eacute;valuations conditionnelles en utilisant la commande elif :</p>

```python
# Méthode interactive
# Le script attendra l'entrée faite par l'usager
#
i = input("Entrer un entier i entre 0 et 100 :")
# Pourquoi le typecasting?
if (int(i)<50):
    print("i est inférieur à 50")
elif (int(i)<25):
    print("i est inférieur à 25")
elif (int(i)<12):
    print("i est inférieur à 12")
else:
    print("i est supérieur ou égal à 50")
```

<p>Pourquoi le <em>elif</em> pourrait &ecirc;tre utile? Contrairement &agrave; C++ ou Java, Python n'a pas de commandes <em>case</em> ou <em>switch</em>, souvent utilis&eacute;s pour passer au travers d'une s&eacute;rie de conditions lors de l'ex&eacute;cution d'un script. Dans ce contexte, <em>elif</em> peut jou&eacute; ce r&ocirc;le. Dans certains cas, selon le type de variables &agrave; &eacute;valuer, on peut profiter d'un it&eacute;rateur implicit:</p>

```python
# Imaginons une variable acquise en amont dans l'éxecution, variable faisant 
# partie d'une structure de données
info = -2

if info in [-3,-1,0,2,4,7] :
    print ('info', info, 'existe')
else :
    print ('info', info, "n'existe pas")
```
