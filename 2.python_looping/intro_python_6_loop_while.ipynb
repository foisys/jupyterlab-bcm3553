<p>Les boucles sont des &eacute;l&eacute;ments incontornables dans le processus de traitement des donn&eacute;es dans le contexte de l'ex&eacute;cution d'un script. Comme tout autre language, Python poss&egrave;de divers m&eacute;canismes permettant d'ex&eacute;cuter (ou pas) des t&acirc;ches:</p>
<ul>
<li>Les boucles <em>if</em>;</li>

<h2 id="Les-boucles-If">Les boucles While</h2>
<p>Les boucles while sont un peu l'antith&egrave;se des boucles if: tant et aussi longtemps qu'une ocndition est respect&eacute;e alors ex&eacute;cuter les commandes qui suivent. Comme la plupart du temps, la condition initiale est vraie (true), &ccedil;a assure que les commandes dans la boucle seront ex&eacute;cut&eacute;es au moins une fois.</p>

```python
# Spécifions une variable avec une valeur initiale
i = 1
message = "0"
while i < 6:
  message = message + "..."+str(i)
  i += 1
print(message)
```

<p>&Agrave; tout moment, &agrave; l'atteinte d'une certaine condition, une boucle <em>while</em> peut &ecirc;tre stopp&eacute;e gr&acirc;ce &agrave; la commande <em>break</em>:</p>

```python
# Spécifions une variable avec une valeur initiale
i = 1
message = "0"
while i < 6:
#  print(i)
  message = message + "..."+str(i)
  i += 1
  if (i > 3):
    break
print(message)
```
