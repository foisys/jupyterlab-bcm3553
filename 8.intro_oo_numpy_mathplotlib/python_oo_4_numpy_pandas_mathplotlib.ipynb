## Représentation graphique de l'exécution d'équations matéhmatiques

Faire des simulations employant des formules mathématiques avec changement de variables est simple!

```python
import numpy as np
import matplotlib.pyplot as mpl

#
# Utilisons numpy pour générer une suite de valeurs entre deux bornes
#
x1 = np.arange(0,5*np.pi-5,0.1)   # start,stop,step

y1 = np.sin(x1)
y2 = np.cos(x1)

mpl.plot(x1,y1)
mpl.plot(x1,y2)

mpl.xlabel('Valeurs de x entre 0 to 5pi')  # string must be enclosed with quotes '  '
mpl.ylabel('sin(x) et cos(x)')
mpl.title('Graphe des valeurs de sin et cos entre 0 et 5pi')
mpl.legend(['sin', 'cos'])      # legend entries as seperate strings in a list

# Sauvegarder la figure
mpl.savefig("../z.misc_files/demo_matplotlib.png")
mpl.show()
```
