<h1 id="Pr&eacute;paration-de-l'environnement-de-d&eacute;veloppement">Pr&eacute;paration de l'environnement de d&eacute;veloppement</h1>
<h2 id="Introduction">Introduction</h2>
<p>Pour r&eacute;soudre notre probl&egrave;me, on a besoin d'un outil: un language de programmation. Ce language doit nous permettre de communiquer avec le mat&eacute;riel de l'ordinateur afin d'ex&eacute;cuter les t&acirc;ches n&eacute;cessaires pour accomplir nos t&acirc;ches: lecture et &eacute;criture sur le syst&egrave;me de fichier, traitement de l'information, connexions avec le monde ext&eacute;rieur si n&eacute;cessaire.</p>
<h2 id="Nos-outils">Nos outils</h2>
<p>Comment &nbsp;utiliser un language? La mani&egrave;re la plus simple se fait via un simple fichier en format texte, &eacute;crit via un &eacute;diteur de texte. Simple mais limit&eacute;... Une approche plus payante est l'utilisation d'un environnement de d&eacute;veloppement int&eacute;gr&eacute;. Duh? C'est un logiciel offrant des fonctionnalit&eacute;s compl&eacute;mentaires comme l'indentation automatique ou l'autocorrection. Il existe une multitude de ces outils mais un d'entre eux qui est tr&egrave;s populaire est <a href="http://jupyter.org" target="_blank" rel="noopener">Jupyter</a>. &nbsp;Pourquoi? Parce qu'en plus d'&eacute;crire du code, il est facile de compl&eacute;ter ce code avec de la documentation dans la m&ecirc;me plateforme.</p>
<h2 id="Installation">Installation</h2>
<p>Donc comment faire avec Jupyter?</p>
<ol>
<li>Premi&egrave;rement, il fait avoir Python 3 install&eacute; sur votre ordinateur. Sur une Mac ou sur un poste Linux/Unix, c'est d&eacute;j&agrave; fait ;-) Sur Windows, il faut installer le package provenant du site de Python: <a title="Page des installateurs Windows 7 et plus" href="https://www.python.org/downloads/windows/" target="_blank" rel="noopener">Page des installateurs Windows 7 et plus.</a></li>
<li>Il est tr&egrave;s facile alors d'insteller Jupyter Lab:</li>
</ol>
<p><code>% pip3 install jupyterlab</code><br /><br />On d&eacute;marre alors notre environement ainsi:</p>
<p><code>% mkdir mon_projet</code><br /><code>% jupyter lab --notebook-dir= ./mon_projet</code></p>
<p>Une fen&ecirc;tre du fureteur s'ouvrira et on peut commencer!<code></code></p>
