<h2 id="Les-fonctions:-automatiser-les-op&eacute;rations-r&eacute;currentes">Introduction au concept des fonctions: automatiser les op&eacute;rations r&eacute;currentes</h2>
<p>Lorsqu'on d&eacute;veloppe des solutions destin&eacute;es &agrave; une certaine permanence ou bien lorsqu'on &eacute;crit script apr&egrave;s script des routines qui utilisent des logiques r&eacute;currentes, il devient tr&egrave;s utile d'utiliser des fonctions. Un autre avantage est de permettre une modularisation de l'ex&eacute;cution d'un script, rendant plus facile son d&eacute;boguage.</p>
<p>On peut &eacute;crire une fonction &agrave; tout endroit d'un script MAIS, et &ccedil;a peut paraitre &eacute;vident, il ne faut jamais mettre la fonction <strong>apr&egrave;s</strong> son invocation dans le script.</p>

```python
#
# On peut mettre des parametres pour etre traités par la fonction
# entre les parenthèses
#
def uneFonction():
    print("Ca marche!!")

uneFonction()
```

```python
#
# Ça, ça ne marchera pas
#
uneFonction()

def uneFonction():
    print("Ca marche!!")
```

<h2 id="Comment-&ccedil;a-marche?-Une-fonction-dans-votre-script">Comment &ccedil;a marche? Une fonction dans votre script</h2>
<p>Comme &eacute;crit ci-dessus, simplement &eacute;crire la ou les fonctions selon votre inspiration. Par exemple, v&eacute;rifions la conformit&eacute; d'un fichier au format FASTA.</p>

```python
def checkFasta(aFile):
    allLines = aFile.readlines()
    if ">" in allLines[0]:
        return True
    else:
        return False

aFile = open("../z.misc_files/h.sapiens.cttn.iso.a.aa.fa","r")

if checkFasta(aFile):
    print("Le fichier est dans le bon format!")
else:
    print("Le fichier n'est pas dans le bon format!")
```

<h2 id="Comment-&ccedil;a-marche?-Une-fonction-dans-votre-script">Comment &ccedil;a marche? Une fonction dans un script externe</h2>
<p>Sommes-nous oblig&eacute; d'&eacute;crire une fonction dans le fichier qui l'ex&eacute;cute? Non :-) C'est tr&egrave;s utile de mettre des fonctions passe-partout dans un script externe qu'on appelle lorsque n&eacute;cessaire dans un autre script qui en a besoin. Imaginons que nous avons un script appel&eacute; fastaTools.py qui se trouve au m&ecirc;me niveau dans le syst&egrave;me de fichiers. fastaTools.py contient une s&eacute;rie de fonctions qui servent &agrave; n'importe quel autre script &agrave; g&eacute;rer les s&eacute;quences FASTA.</p>

```python
#
# Vérifions le contenu du fichier fastaTools
#
!less fastaTools.py
```

```python
#
# Utilisons maitenant la fonction dans notre cellule de code
#
# On importe toutes les fonctions qui seraient contenu dans fastaTools
import fastaTools as f

aFile = open("../z.misc_files/h.sapiens.cttn.iso.a.cdna.bad.fa","r")

#
# Il faut metttre le raccourci "f" pour que le script fonctionne
# ou bien "fastaTools."
#
if f.checkFasta(aFile):
    print("Le fichier est dans le bon format!")
else:
    print("Le fichier n'est pas dans le bon format!")
```

```python
#
# Utilisons maitenant la fonction dans notre cellule de code
#
# Si on ne veut que quelques fonctions, on les appelle individuellement:
from fastaTools import checkFasta

aFile = open("../z.misc_files/h.sapiens.cttn.iso.a.aa.fa","r")

if checkFasta(aFile):
    print("Le fichier est dans le bon format!")
else:
    print("Le fichier n'est pas dans le bon format!")
```
