# Comment faire une représentation graphique des annotations d'une séquence

## Introduction

Dans plusieurs situations (publications ou pages web), on dispose d'une séquence d'ADN ou de protéines d'un côté et des éléments descriptifs ou fonctionnels (*features*) d'un autre côté et on veut fusionner les deux ensembles de données dans un graphique. Comment faire? En utilisant la librairie `GenomeDiagram` incluse avec BioPython.

#### Pré-requis nécessaires

Avant de commencer, assurons-nous que les modules nécessaires soient installés sur notre plateforme ;-) Pour certaines fonctions de `GenomeDiagram`, les packages `ReportLab` et `Python Imaging Library` sont nécessaires en plus. 

NOTE: sur une serveur de l'environnement SENS, ça ne fonctionnera pas... Il vous faut passer par un environnement virtuel (*virtualenv*) créé dans votre espace perso.

#### Code de démonstration

Voici notre exemple avec le plasmide pGEM-9Zf.

```python
from reportlab.lib import colors
from reportlab.lib.units import cm
from Bio.Graphics import GenomeDiagram
from Bio import SeqIO

#
# Lecture d'un fichier de sequence avec annotations en format Genbank
# Pour un fichier sans annotations (ie: format FASTA), ça demandera un 
# peu plus de travail... Plus à venir :-)
#
uneSeq = SeqIO.read("../z.misc_files/data_seq/vector.pGEM-9Zf-minus.gb", "genbank")

#
# ajustons certains parametres et créons un diagramme vide pour commencer.
#
gen_diagram = GenomeDiagram.Diagram("Carte des attributs sur la séquence pGEM-9Zf(-)")
gen_track_for_features = gen_diagram.new_track(1, name="attributs")
gen_feature_set = gen_track_for_features.new_set()

#
# Mettons en évidence tout attribut avec un label "gene"
#
for attribut in uneSeq.features:
    if attribut.type != "gene":
        # On n'en veut pas :-)
        continue
    if len(gen_feature_set) % 2 == 0:
        couleur = colors.blue
    else:
        couleur = colors.lightblue
    gen_feature_set.add_feature(attribut, color=couleur, label=True)

#
# Il faut maintenant construire notre graphique 
#
gen_diagram.draw(
    format="circular",
    circular=True,
    orientation="portrait",
    # La methode draw() utilise les valeurs ISO pas US :-(
    # mais accepte les valeurs en metrique
    pagesize=(20 * cm, 20 * cm),
    # La portion devant rester libre
    circle_core=0.9,
    start=0,
    end=len(uneSeq),
)

#
# On peut aussi faire d'autres formats comme SVG ou PNG en ajustant le 
# 2eme parametre de la fonction write()
#
gen_diagram.write("../z.misc_files/data_seq/vector.pGEM-9Zf-minus.pdf", "PDF")

#
# Plus à venir...
#
```
