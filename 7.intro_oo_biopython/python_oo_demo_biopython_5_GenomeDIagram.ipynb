# Comment faire une reprГ©sentation graphique des annotations d'une sГ©quence

## Introduction

Dans plusieurs situations (publications ou pages web), on dispose d'une sГ©quence d'ADN ou de protГ©ines d'un cГґtГ© et des Г©lГ©ments descriptifs ou fonctionnels (*features*) d'un autre cГґtГ© et on veut fusionner les deux ensembles de donnГ©es dans un graphique. Comment faire? En utilisant la librairie `GenomeDiagram` incluse avec BioPython; sa documentation se trouve dans la documentation gГ©nГ©rale de BioPython.

#### Code de dГ©monstration

Voici notre exemple avec le plasmide pGEM-9Zf.

```python
from reportlab.lib import colors
from reportlab.lib.units import cm
from Bio.Graphics import GenomeDiagram
from Bio import SeqIO

#
# Lecture d'un fichier de sequence avec annotations en format Genbank
# Pour un fichier sans annotations (ie: format FASTA), Г§a demandera un 
# peu plus de travail... Plus Г  venir :-)
#
uneSeq = SeqIO.read("../z.misc_files/data_seq/vector.pGEM-9Zf-minus.gb", "genbank")

#
# ajustons certains parametres et crГ©ons un diagramme vide pour commencer.
#
gen_diagram = GenomeDiagram.Diagram("Carte des attributs sur la sГ©quence pGEM-9Zf(-)")
gen_track_for_features = gen_diagram.new_track(1, name="attributs")
gen_feature_set = gen_track_for_features.new_set()

#
# Mettons en Г©vidence tout attribut avec un label "gene"
#
for attribut in uneSeq.features:
    if attribut.type != "gene":
        # On n'en veut pas :-)
        continue
    if len(gen_feature_set) % 2 == 0:
        couleur = colors.blue
    else:
        couleur = colors.lightblue
    gen_feature_set.add_feature(attribut, color=couleur, label=True)

#
# Il faut maintenant construire notre graphique 
#
gen_diagram.draw(
    format="circular",
    circular=True,
    orientation="portrait",
    # La methode draw() utilise les valeurs ISO pas US :-(
    # mais accepte les valeurs en metrique
    pagesize=(20 * cm, 20 * cm),
    # La portion devant rester libre
    circle_core=0.9,
    start=0,
    end=len(uneSeq),
)

#
# On peut aussi faire d'autres formats comme SVG ou PNG en ajustant le 
# 2eme parametre de la fonction write()
#
gen_diagram.write("../z.misc_files/data_seq/vector.pGEM-9Zf-minus.pdf", "PDF")

#
# Plus Г  venir...
#
```
