<p>La plus grande partie du travail que nous faisons n&eacute;cessite l'emploi de fichiers: on doit les lire, on doit en &eacute;crire, on doit en modifier. Comment faire en Python? Ce n'Est pas tr&egrave;s compliqu&eacute; mais avant tout, il faut connaitre quelques rudiments de gestion d'un syst&egrave;me de fichiers.</p>
<h2>Qu'est-ce qu'un syst&egrave;me de fichiers?</h2>
<p>Tous les syst&egrave;mes en un un et certains (dans l'univers UNIX/Linux) peuvent en avoir plusieurs! C'est essentiellement la portion du syst&egrave;me d'op&eacute;ration de l'ordinteur qui g&egrave;re les &eacute;l&eacute;ments d'information contenu sur l'espace de stockage de l'ordinateur. Dans notre cas, le concept le plus important &agrave; comprendre est la notion des permissions: qui a le droit de faire quoi sur un ordinateur. Sous Linux, dans sa version la plus simple, chaque usager poss&egrave;de un espace usager (son $HOME) o&ugrave; il peut faire pas mal tout ce qu'il veut. En plus, chacun fait partie d'un groupe d'usagers pouvant se partager de l'information et tous les usagers de tous les groupes sont aussi regroup&eacute;s dans la cat&eacute;gorie <em>other</em>.</p>
<p>Voici un exemple:</p>

```python
!ls -l
```

<p>Vous voyez le groupe de lettres et de tirets sous "total"? Ce sont les permissions pour les fichiers contenus dans le r&eacute;pertoire pr&eacute;sent. Donc:</p>
<ul>
<li>Le premier tiret nous dit que le fichier python_files_1.ipynb est un fichier (duh!); si on voyait un d, &ccedil;a nous dirait que ce serait un r&eacute;pertoire (un r&eacute;pertoire est un fait un fichier contenant des pointeurs vers un ensemble de fichiers: son contenu), si on voyait un l, &ccedil;a nous dirait que c'est un lien symbolique (AKA un alias).</li>
<li>Suit ensuite un groupe de 9 symboles, divis&eacute; en 3 sous-groupes de 3: ce sont, de la gauche vers la droite, les permissions pour l'usager (groupe 1), le groupe (groupe 2) et tous les autres usagers (groupe 3).</li>
<li>Le reste qui est important est le nom de l'usager (foisys) et son groupe (staff, par d&eacute;faut le groupe de tous les usagers dans Mac OS X).</li>
<li>Les lettres dans chaque groupe repr&eacute;sente les permissions de lecture (r), d'&eacute;criture (w) ou d'ex&eacute;cution (x); on r&eacute;sum&eacute; noter fichier peut &ecirc;tre lu et modifi&eacute; par moi et lu par tous. Si un fichier avait un x comme permission, &ccedil;a voudrait dire qu'il est ex&eacute;cutable: c'est un programme!</li>
</ul>
