<h1 id="Aspects-(plus)-avanc&eacute;s-pour-la-bioinformatique:-recherche-dans-une-base-de-donn&eacute;es-distante,-version-JSON-et-MyGene.info">Aspects (plus) avanc&eacute;s pour la bioinformatique: ex&eacute;cution d'un programme &agrave; distance via un URL</h1>
<p>Il est aussi possible de cr&eacute;er un URL qui sera utilis&eacute; comme ligne de commande dans l'ex&eacute;cution d'un programme r&eacute;sidant sur un serveur distant via l'Internet. COmment &ccedil;a marche? Un service "&eacute;coute" les requ&ecirc;tes qui arrivent au serveur pour ensuite les utiliser dans l'ex&eacute;cution d'un programme sp&eacute;cifique sur le serveur.</p>
<h2>Un exemple: BLAST au NCBI</h2>
<p>le logiciel d'alignement de s&eacute;quences pairwise NCBI BLAST est, et de loin, le logiciel le plus utilis&eacute; en biologie mol&eacute;culaire. Rappel: une s&eacute;quence inconnue (contenant des nucl&eacute;otides ou des acides amin&eacute;s) est compar&eacute;e contre toutes les s&eacute;quences contenues dans une base de donn&eacute;es; suite &agrave; l'ex&eacute;cution de BLAST, celui-ci retourne les r&eacute;sultats de la recherche de similarit&eacute; entre l'inconnue et des s&eacute;quences connues.</p>
<p>&nbsp;</p>

```python
import requests

# Demandé par le NCBI pour faire le suivi des requetes
email = "sylvain.foisy@umontreal.ca"
tool = "bcm3553_demo_blast"

#
# Consultez la documentation NCBI pour modifier selon vos besoins
#
# La liste des paramétres BLAST utilisables est ici:
# https://ncbi.github.io/blast-cloud/dev/api.html
#
blastTool = "blastp"
blastDb = "nr"

#
# Lire un fichier FASTA, ici de protéines
# On se casse pas la tête avec la vérification...
#
aFile = open("../z.misc_files/h.sapiens.cttn.iso.a.aa.fa","r")
seq = ""
allLines = aFile.readlines()
for aLine in allLines:
    if aLine[0]!='>':
        aLine  = aLine.replace('\n','')
        seq  = seq+aLine

cmds = "CMD=Put&email={}&tool={}&PROGRAM={}&DATABASE={}&QUERY={}".format(email,tool,blastTool,blastDb,seq)
#print(cmds)
# Preparation de la requete
headers = {'content-type': 'application/x-www-form-urlencoded','connection':'keep-alive'}
blast_url = "https://blast.ncbi.nlm.nih.gov/Blast.cgi?"+cmds
# Envoyez la requete
res = requests.post(blast_url,headers=headers)
print(res.status_code)
#
# Work in progress: trouver le moyen de parser la sortie pour intercepter le RID
#
#print(res.text)

#
# Une fois que l'analyse est complétée, utiliser le RID pour récupérer vos résultats d'a;ignement
#
out = requests.get("https://blast.ncbi.nlm.nih.gov/blast/Blast.cgi?CMD=Get&FORMAT_TYPE=Text&RID=UDVW064W013",headers=headers)

print(out.text)
```
